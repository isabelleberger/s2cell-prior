## Prep yeast data

Using yeast data already compiled: https://github.com/Roy-lab/merlin-p_inferred_networks/tree/master/yeast_networks


```python
import pandas as pd
import urllib
import os
```

```python
#Download priors
PATH = '../data/yeast_networks/priors/'
DWNLD = 'https://github.com/Roy-lab/merlin-p_inferred_networks/raw/master/yeast_networks/priors/'
files = ['ChIP_prior.txt', 'KO_prior.txt','Motif_prior.txt', ]
os.makedirs(PATH, exist_ok=True)
for file in files:
    urllib.request.urlretrieve(DWNLD+file, PATH+file)
```

```python
#Download gold std
PATH = '../data/yeast_networks/gold/'
DWNLD = 'https://github.com/Roy-lab/merlin-p_inferred_networks/raw/master/yeast_networks/gold/'
files = ['MacIsaac2.KO.txt','MacIsaac2.NatVar.txt','MacIsaac2.Stress.txt','YEASTRACT_Count3.KO.txt',
         'YEASTRACT_Count3.NatVar.txt','YEASTRACT_Count3.Stress.txt','YEASTRACT_Type2.KO.txt',
         'YEASTRACT_Type2.NatVar.txt', 'YEASTRACT_Type2.Stress.txt']
os.makedirs(PATH, exist_ok=True)
for file in files:
    urllib.request.urlretrieve(DWNLD+file, PATH+file)
```

```python
#Download expression data
PATH = '../data/yeast_networks/expression/'
DWNLD = 'https://github.com/Roy-lab/merlin-p_inferred_networks/raw/master/yeast_networks/expression/'
files = ['KO.txt.tar.gz', 'KO_TF_names.txt','NatVar.txt.tar.gz','NatVar_TF_names.txt','Stress.txt.tar.gz',
         'Stress_TF_names.txt']
os.makedirs(PATH, exist_ok=True)
for file in files:
    urllib.request.urlretrieve(DWNLD+file, PATH+file)
#use tar -zxvf to uncompress tar.gz
```
