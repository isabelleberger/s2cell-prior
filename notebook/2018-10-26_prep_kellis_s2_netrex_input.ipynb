## run NetREX 1 with Kellis prior

- Need updated S2 cell expression matrix from Justin
- Need to download and import/ look at Kellis prior

```python
import pandas as pd
```

```python
#read expression matrix

rpkm_cnts = pd.read_table('../output/rna_seq/s2_rnaseq_rpkm_20181025.tsv', index_col=0)
rpkm_cnts.head()
```

```python
filter_rpkm = rpkm_cnts[rpkm_cnts.max(axis=1) > 4]
```

```python
filter_rpkm.shape
```

```python
#read supervised FlyNet prior from Marbach et al

prior = pd.read_table('../data/flynet_supervised_0.6.txt', header=None, names=['TF','gene', 'weight'])
prior.head()
```

```python
matrix_prior = prior.set_index(['gene','TF']).unstack().fillna(value=0)
```

```python
matrix_prior.shape
```

```python
good_idx = filter_rpkm.index.intersection(matrix_prior.index)
```

```python
idx_rpkm = filter_rpkm.loc[good_idx]
idx_prior = matrix_prior.loc[good_idx]
```

```python
print(idx_rpkm.shape, idx_prior.shape)
```

```python
idx_rpkm.to_csv('../output/s2cell_rpkm_idx_for_netrex', sep='\t', header=None)
```

```python
idx_prior.columns = idx_prior.columns.get_level_values(1)
idx_prior.index.name = None
idx_prior.columns.name = None
idx_prior.head()
```

```python
trim_zeros_prior = idx_prior.loc[:, (idx_prior != 0).any(axis=0)]
```

```python
trim_zeros_prior.to_csv('../output/kellis_prior_idx_for_netrex', sep='\t')
```
