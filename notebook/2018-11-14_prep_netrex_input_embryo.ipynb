# Prep embryo netrex input 

```python
import pandas as pd
```

```python
rpkm_cnts = pd.read_table('https://hpc.nih.gov/~fearjm/s2rnai/2018-11-13_embryo_rpkm.tsv', index_col=0)
```

```python
filtered = rpkm_cnts[rpkm_cnts.max(axis=1) > 4]
```

```python
filtered.shape
```

```python
filtered.head()
```

```python
#read supervised FlyNet prior from Marbach et al

prior = pd.read_table('../data/flynet_supervised_0.6.txt', header=None, names=['TF','gene', 'weight'])
prior.head()
```

```python
matrix_prior = prior.set_index(['gene','TF']).unstack().fillna(value=0)
```

```python
good_idx = filtered.index.intersection(matrix_prior.index)

idx_rpkm = filtered.loc[good_idx]
idx_prior = matrix_prior.loc[good_idx]

print(idx_rpkm.shape, idx_prior.shape)
```

```python
idx_prior.columns = idx_prior.columns.get_level_values(1)
idx_prior.index.name = None
idx_prior.columns.name = None
idx_prior.head()
```

```python

trim_zeros_prior = idx_prior.loc[:, (idx_prior != 0).any(axis=0)]
```

```python
trim_zeros_prior.shape
```

```python
idx_rpkm.to_csv('../output/embryo_rpkm_idx_for_netrex', sep='\t', header=None)
trim_zeros_prior.to_csv('../output/embryo_prior_idx_for_netrex', sep='\t')
```
