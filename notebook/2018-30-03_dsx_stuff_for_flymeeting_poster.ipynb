# DSX FOR POSTER

```python
import pandas as pd
import gffutils
from gffutils import pybedtools_integration
import pybedtools
from pybedtools.featurefuncs import gff2bed
```

```python
peaks_ = pd.read_table('../output/chip/dsx_peaks.bed')
```

```python
peaks_.head()
```

```python
#Target gene intersect

gene_info = pybedtools.BedTool('../output/chip/dmel6.12.genes.bed')

targene_intersect = gene_info.intersect(pybedtools.BedTool.from_dataframe(peaks_), 
                                        wb=True).saveas().to_dataframe()[[3,6,7,8,9,10,11,12,13]]

targene_intersect.columns = ['target_gene','chrom','start','end','srx','log10qval','caller','TF','TF_fbgn']

targene_intersect.head()
```

```python
#make a simple bed for IGV: 

bed = targene_intersect[['chrom','start','end','TF','log10qval']]
```

```python
bed['strand'] = '.'
bed.drop_duplicates(inplace=True)
```

```python
bed.to_csv('../output/chip/dsx__igv.bed', header=None, index=False, sep='\t')
```
